```python
url = 'https://www.terra.com.br/esportes/futebol/brasileiro-serie-a/tabela/'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'}
site = requests.get(url, headers=headers)

dados = pd.read_html(site.text)

df = pd.DataFrame(dados[0])

df
```

```python
df.drop(['Times', 'Times.1'], axis=1, inplace=True)
df.info()
```

```python
df
```

```python
df.rename(columns={'Times.2': 'Times'}, inplace=True)
df
```

```python
# Salvando em excel

excel = pd.ExcelWriter('BRASILEIRO 22.xlsx', engine='xlsxwriter')
df.to_excel(excel, sheet_name='BR22')
excel.save()
```
